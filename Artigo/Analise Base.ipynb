## Análise dos dados:
<ul>
    <li> datas_dados: tabela com número de tweets e retweets, absoluto e relativos da base de dados por dia</li>
    <li> datas_original: abela com número de tweets e retweets do arquivo original por dia</li>
    <li> dados: tabela com número de tweets, datas e quantidade de dias em que aparece, por usuário</li>

    <li> tts: número de tweets na base</li>
    <li> rtts_original: número de retweets no arquivo original</li>
    <li> tts_original: número de tweets no arquivo original</li>
    <li> usuários: número de usuários na base</li>
</ul>

```python
import pandas as pd
```

```python
base = pd.read_csv('cont_usuarios_data_100k_origin_teste.csv',sep=';')
dados = base.groupby('Usuários').agg({'Tweets':sum,'Datas':list}).sort_values('Tweets',ascending=False)
dias = base.groupby('Usuários').agg({'Tweets':sum,'Datas':list}).sort_values('Tweets',ascending=False)['Datas'].agg(len)
```

```python
dados['Dias'] = dias
dados
```
